```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

class Customer:
    def __init__(self, name, identity, phone, email, rank):
        self.name = name
        self.identity = identity
        self.phone = phone
        self.email = email
        self.rank = rank
    def __str__(self):
        return f'Họ tên: {self.name}, Mã khách hàng: {self.identity}, SĐT: {self.phone}, Email: {self.email}, Hạng: {self.rank}'

window = Tk()
window.title('Customer Management System')
window.geometry("1000x550")
window.iconbitmap(r'D:\PYF2304R1\BasicPythonCodeGym\Case Study\crm.ico')

style = ttk.Style()
style.theme_use('xpnative')
style.configure('Treeview', background = '#D3D3D3', foreground = 'black', rowheight = 25, fieldbackground = '#D3D3D3')
style.map('Treeview', background = [('selected', '#347083')])

tree_frame = Frame(window)
tree_frame.pack(pady = 10)

tree_scroll = Scrollbar(tree_frame)
tree_scroll.pack(side=RIGHT, fill = Y)

tree = ttk.Treeview(tree_frame, yscrollcommand = tree_scroll.set, selectmode = 'extended')
tree.pack()

tree_scroll.config(command = tree.yview)

tree['columns'] = ('Họ tên', 'Mã khách hàng', 'Số điện thoại', 'Email', 'Hạng')
tree.column('#0', width = 0, stretch = NO)
tree.column('Họ tên', anchor = W, width = 140)
tree.column('Mã khách hàng', anchor = CENTER, width = 100)
tree.column('Số điện thoại', anchor = CENTER, width = 140)
tree.column('Email', anchor = CENTER, width = 140)
tree.column('Hạng', anchor = CENTER, width = 140)

tree.heading('#0', text = '', anchor = W)
tree.heading('Họ tên', text = 'Họ tên', anchor = W)
tree.heading('Mã khách hàng', text = 'Mã khách hàng', anchor = CENTER)
tree.heading('Số điện thoại', text = 'Số điện thoại', anchor = CENTER)
tree.heading('Email', text = 'Email', anchor = CENTER)
tree.heading('Hạng', text = 'Hạng', anchor = CENTER)

customer_list = []

tree.tag_configure('oddrow', background = 'white')
tree.tag_configure('evenrow', background = 'lightblue')



data_frame = LabelFrame(window, text = 'Thông tin khách hàng')
data_frame.pack(fill = 'x', expand = 'yes', padx = 20)

lbl_name = Label(data_frame, text = 'Họ tên')
lbl_name.grid(row = 0, column = 0, padx = 10, pady = 10)
ent_name = Entry(data_frame)
ent_name.grid(row = 0, column = 1, padx = 10, pady = 10)

lbl_id = Label(data_frame, text = 'Mã khách hàng')
lbl_id.grid(row = 0, column = 2, padx = 10, pady = 10)
ent_id = Entry(data_frame)
ent_id.grid(row = 0, column = 3, padx = 10, pady = 10)

lbl_phone = Label(data_frame, text = 'Số điện thoại')
lbl_phone.grid(row = 1, column = 0, padx = 10, pady = 10)
ent_phone = Entry(data_frame)
ent_phone.grid(row = 1, column = 1, padx = 10, pady = 10)

lbl_email = Label(data_frame, text = 'Email')
lbl_email.grid(row = 1, column = 2, padx = 10, pady = 10)
ent_email = Entry(data_frame)
ent_email.grid(row = 1, column = 3, padx = 10, pady = 10)

lbl_rank = Label(data_frame, text = 'Hạng')
lbl_rank.grid(row = 1, column = 4, padx = 10, pady = 10)
ent_rank = Entry(data_frame)
ent_rank.grid(row = 1, column = 5, padx = 10, pady = 10)

count = 0

def add_customer():
    name_value = ent_name.get()
    identity_value = ent_id.get()
    phone_value = ent_phone.get()
    email_value = ent_email.get()
    rank_value = ent_rank.get()
    
    check = False
    
    if name_value == '' or identity_value == '' or phone_value == '':
        messagebox.showerror('Lỗi thông tin bắt buộc', 'Tên, mã khách hàng và số điện thoại là thông tin bắt buộc')
        check = True
    elif identity_value.startswith('0'):
        messagebox.showerror('Lỗi mã khách hàng', 'Mã khách hàng không được bắt đầu với 0')
        check = True
    else:        
        for i in range (len(customer_list)):
            if customer_list[i].phone == phone_value or customer_list[i].identity == identity_value:
                messagebox.showwarning('Đã có khách hàng trong dữ liệu', f'Trùng mã khách hàng hoặc số điện thoại, đã có mã khách hàng hoặc số điện thoại trong cơ sở dữ liệu')
                check = True
                break
    
    if check == False:    
        if email_value == '':
            email_value = 'Không cung cấp'
        if rank_value == '':
            rank_value = 'Thành viên'
        new_customer = Customer(name_value, identity_value, phone_value, email_value, rank_value)
        customer_list.append(new_customer)
        global count
        if count%2 == 0:
            tree.insert(parent = '', index = 'end', iid = count, text = '', values = (new_customer.name, new_customer.identity, new_customer.phone, new_customer.email, new_customer.rank), tags = ('evenrow',))
        else:
            tree.insert(parent = '', index = 'end', iid = count, text = '', values = (new_customer.name, new_customer.identity, new_customer.phone, new_customer.email, new_customer.rank), tags = ('oddrow',))
        count += 1
        messagebox.showinfo('Thêm khách hàng thành công', f'Đã thêm khách hàng {new_customer}')
        
        ent_name.delete(0, END)
        ent_id.delete(0, END)
        ent_phone.delete(0, END)
        ent_email.delete(0, END)
        ent_rank.delete(0, END)
        
def move_row_up():
    rows = tree.selection()
    if rows == ():
        messagebox.showerror('Chưa chọn thông tin', 'Vui lòng chọn thông tin khách hàng để di chuyển')
    else:
        for row in rows:
            tree.move(row, tree.parent(row), tree.index(row)-1)

def move_row_down():
    rows = tree.selection()
    if rows == ():
        messagebox.showerror('Chưa chọn thông tin', 'Vui lòng chọn thông tin khách hàng để di chuyển')
    else:
        for row in reversed(rows):
            tree.move(row, tree.parent(row), tree.index(row)+1)
        
def select_record(e):
    
    ent_name.delete(0, END)
    ent_id.delete(0, END)
    ent_phone.delete(0, END)
    ent_email.delete(0, END)
    ent_rank.delete(0, END)
    
    selected = tree.focus()
    if selected == '':
        print('Chưa chọn thông tin khách hàng')
    else:
        values = tree.item(selected, 'values')
    
        ent_name.insert(0, values[0])
        ent_id.insert(0, values[1])
        ent_phone.insert(0, values[2])
        ent_email.insert(0, values[3])
        ent_rank.insert(0, values[4])

def clear_entries():
    
    ent_name.delete(0, END)
    ent_id.delete(0, END)
    ent_phone.delete(0, END)
    ent_email.delete(0, END)
    ent_rank.delete(0, END)
    
    
def delete_one():
    
    selected = tree.focus()
    values = tree.item(selected)
    if selected == '':
        messagebox.showerror('Chưa chọn thông tin', 'Vui lòng chọn thông tin khách hàng để xóa')
    else:
        delete_id = values.get('values')[1]
        check = False
        for i in range(len(customer_list)):
            if customer_list[i].identity == delete_id:
                del customer_list[i]
                tree.delete(selected)
                messagebox.showinfo('Đã xóa thành công', 'Đã xóa thông tin khách hàng thành công')
                check = True
                
                ent_name.delete(0, END)
                ent_id.delete(0, END)
                ent_phone.delete(0, END)
                ent_email.delete(0, END)
                ent_rank.delete(0, END)
                
                break
    
        if check == False:
            messagebox.showerror('Không xóa thành công', '''Không tìm thấy ID để xóa, chọn lại thông tin khách hàng. **LƯU Ý: Mã KH không được bắt đầu bằng số 0**''')
    

def delete_all():
    
    for record in tree.get_children():
        tree.delete(record)
    customer_list.clear()
        
def update_record():
    
    selected = tree.focus()
    values = tree.item(selected)
    if selected == '':
        messagebox.showerror('Chưa chọn thông tin', 'Vui lòng chọn thông tin để cập nhật')
    
    else:
        updated_id = values.get('values')[1]
        updated_name = ent_name.get()
        updated_phone = ent_phone.get()
        updated_email = ent_email.get()
        updated_rank = ent_rank.get()
        
        check = False
        if updated_name == '' or updated_id == '' or updated_phone == '' or updated_email == '' or updated_rank == '':
            messagebox.showerror('Thiếu thông tin cập nhập', 'Vui lòng cập nhập đủ thông tin')
        elif updated_id != ent_id.get():
            messagebox.showerror('Lỗi mã khách hàng', 'Vui lòng không sửa mã khách hàng')
        else:
            for i in range (len(customer_list)):
                if customer_list[i].identity == updated_id:
                    customer_list[i].name = updated_name
                    customer_list[i].phone = updated_phone
                    customer_list[i].email = updated_email
                    customer_list[i].rank = updated_rank
                
                    tree.item(selected, text = '', values = (customer_list[i].name, customer_list[i].identity, customer_list[i].phone, customer_list[i].email, customer_list[i].rank))

                    ent_name.delete(0, END)
                    ent_id.delete(0, END)
                    ent_phone.delete(0, END)
                    ent_email.delete(0, END)
                    ent_rank.delete(0, END)
                    messagebox.showinfo('Cập nhập thành công', 'Thông tin khách hàng đã được cập nhập')
                    
                    check = True
                    break
            if check == False:
                    messagebox.showerror('Không tìm thấy mã khách hàng', 'Không tìm thấy mã khách hàng cần cập nhập. **LƯU Ý** Mã Khách Hàng không được bắt đầu bằng số 0')

def search_customer():
    searched_value = ent_search.get()
    check = False
    
    for i in range(len(customer_list)):
        if customer_list[i].identity == searched_value or customer_list[i].phone == searched_value:
            messagebox.showinfo('Tìm thấy khách hàng', customer_list[i])
            check = True
            break
    if check == False:
        messagebox.showerror('Không thấy khách hàng', 'Không tìm thấy khách hàng. Vui lòng tìm theo mã khách hàng hoặc số điện thoại khách hàng')
    ent_search.delete(0, END)
    
def instruction():
    messagebox.showinfo('Hướng dẫn sử dụng', '''
    ** HƯỚNG DẪN SỬ DỤNG CÁC LỆNH **
    
    1. THÊM THÔNG TIN:
    Điền thông tin khách hàng vào các ô thông tin ở phần "Thông tin khách hàng" sau đó ấn nút để nhập thông tin khách hàng mới vào cơ sở dữ liệu
    Lưu Ý: 3 thông tin bắt buộc là tên, mã khách hàng và số điện thoại và mã khách hàng không được bắt đầu bằng số 0
    
    2. CẬP NHẬP:
    Ấn vào 1 dòng thông tin khách hàng trên bảng thông tin phía trên, ở phần "Thông tin khách hàng" sẽ hiện ra các thông tin của khách hàng đó
    Sửa các thông tin cần cập nhập rồi ấn nút để cập nhập thông tin vào cơ sở dữ liệu.
    Lưu Ý: Không sửa mã khách hàng, mã khách hàng là duy nhất, nếu mã sai vui lòng xóa khách hàng cũ và thêm lại khách hàng mới
    
    3. XÓA 1 THÔNG TIN:
    Ấn vào 1 dòng thông tin khách hàng trên bảng thông tin phía trên, ở phần "Thông tin khách hàng" sẽ hiện ra các thông tin của khách hàng đó
    Ấn nút để xóa thông tin khách hàng đó
    
    4. XÓA TOÀN BỘ THÔNG TIN:
    Ấn nút để xóa toàn bộ thông tin khách hàng trong cơ sở dữ liệu
    
    5. CHUYỂN THÔNG TIN LÊN:
    Chọn 1 thông tin khách hàng trên bảng thông tin phía trên rồi ấn nút để chuyển thông tin khách được chọn lên 1 vị trí
    
    6. CHUYỂN THÔNG TIN XUỐNG:
    Chọn 1 thông tin khách hàng trên bảng thông tin phía trên rồi ấn nút để chuyển thông tin khách được chọn xuống 1 vị trí
    
    7. LÀM TRỐNG CÁC Ô NHẬP LIỆU:
    Ấn vào nút "Làm trống các ô nhập liệu" để làm trống toàn bộ các ô ở phần "Thông tin khách hàng"
    
    8. TÌM THÔNG TIN KHÁCH HÀNG THEO MÃ HOẶC SĐT:
    Điền mã khách hàng hoặc số điện thoại khách hàng cần tìm vào ô nhập liệu cạnh nút tìm rồi ấn nút để tìm thông tin về khách hàng đó
    ''')

button_frame = LabelFrame(window, text = 'Các lệnh tác nghiệp')
button_frame.pack(fill = 'x', expand = 'yes', padx = 20)

btn_update = Button(button_frame, text = 'Cập nhập', command = update_record)
btn_update.grid(row = 0, column = 1, padx = 10, pady = 10)

btn_add = Button(button_frame, text = 'Thêm thông tin', command = add_customer)
btn_add.grid(row = 0, column = 0, padx = 10, pady = 10)

btn_delete_one = Button(button_frame, text = 'Xóa 1 thông tin', command = delete_one)
btn_delete_one.grid(row = 0, column = 2, padx = 10, pady = 10)

btn_delete_all = Button(button_frame, text = 'Xóa toàn bộ thông tin', command = delete_all)
btn_delete_all.grid(row = 0, column = 3, padx = 10, pady = 10)

btn_instruct = Button(button_frame, text = 'Hướng dẫn sử dụng các lệnh', command = instruction)
btn_instruct.grid(row = 0, column = 4, padx = 10, pady = 10)

btn_move_up = Button(button_frame, text = 'Chuyển thông tin lên', command = move_row_up)
btn_move_up.grid(row = 1, column = 0, padx = 10, pady = 10)

btn_move_down = Button(button_frame, text = 'Chuyển thông tin xuống', command = move_row_down)
btn_move_down.grid(row = 1, column = 1, padx = 10, pady = 10)

btn_clear = Button(button_frame, text = 'Làm trống các ô nhập liệu', command = clear_entries)
btn_clear.grid(row = 1, column = 2, padx = 10, pady = 10)

ent_search = Entry(button_frame)
ent_search.grid(row =1, column = 3, padx = 10, pady = 10)

btn_search = Button(button_frame, text = 'Tìm thông tin khách hàng theo mã hoặc SĐT', command = search_customer)
btn_search.grid(row = 1, column = 4, padx = 10, pady = 10) 

tree.bind('<ButtonRelease-1>', select_record)

window.mainloop()
```
