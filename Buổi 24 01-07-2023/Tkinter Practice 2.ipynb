```python
class Book:
    def __init__(self, name, category, price):
        self.name = name
        self.category = category
        self.price = price
    def __str__(self):
        return f'Name: {self.name}, category: {self.category}, price: {self.price}'
```

```python
from tkinter import *
from tkinter import messagebox

window = Tk()
window.title('Book Management System')

book_list = []

def add_book():
    name_value = (name.get()).lower()
    category_value = (category.get()).lower()
    price_value = price.get()
    check = False
    if name_value == '' or category_value == '' or price_value =='':
        messagebox.showerror('Book Info', 'Please enter name, category and price')
    else:
        for i in range (len(book_list)):
            if book_list[i].name == name_value:
                messagebox.showwarning('Book Already in Database', f'The book: {name_value.upper()} is already in the Database')
                check = True
                break
            
        if check == False:    
            new_book = Book(name_value, category_value, float(price_value))
            book_list.append(new_book)
        
            print([book.name for book in book_list])
            messagebox.showinfo('Added successfully', f'Added the book {name_value.upper()}, category: {category_value.upper()}, price: {price_value}$')

def delete_book():
    delete_name = (search_name.get()).lower()
    check = False
    
    for i in range(len(book_list)):
        if book_list[i].name == delete_name:
            del book_list[i]
            print([book.name for book in book_list])
            messagebox.showinfo('Delete successfully', f'Deleted {delete_name.upper()} successfully')
            check = True
            break
    
    if check == False:
        messagebox.showerror('Cannot Delete', 'Cannot find the book to delete')

def search_book():
    searched_name = (search_name.get()).lower()
    check = False
    
    for i in range(len(book_list)):
        if book_list[i].name == searched_name:
            messagebox.showinfo('Book Found', f'Book Status: Name: {book_list[i].name.upper()}, Category: {book_list[i].category.upper()}, Price: {book_list[i].price}$')
            check = True
            break
    if check == False:
        messagebox.showerror('Book Not Found', 'Cannot find the book')

def reset():
    name.delete(0, 'end')
    category.delete(0, 'end')
    price.delete(0, 'end')
    search_name.delete(0, 'end')

def update_price():
    updated_book = (name.get()).lower()
    updated_price = price.get()
    
    check = False
    if updated_book == '' or updated_price == '':
        messagebox.showerror('Updated Price Error', 'Please enter name and updated price')
    else:
        for i in range (len(book_list)):
            if book_list[i].name == updated_book:
                book_list[i].price = updated_price
                messagebox.showinfo('Price Updated', f'The price of the book {updated_book.upper()} has been updated to {updated_price}$')
                check = True
                break
        if check == False:
                messagebox.showerror('Cannot Find Book', 'Cannot find the book in Database')

lbl_name = Label(window, text = 'Name:')
lbl_category = Label(window, text = 'Category:')
lbl_price = Label(window, text = 'Price')

name = Entry(window)
category = Entry(window)
price = Entry(window)
search_name = Entry(window)

btn_add = Button(window, text ='Add', command = add_book)
btn_delete = Button(window, text = 'Delete', command = delete_book)
btn_reset = Button(window, text = 'Reset', command = reset)
btn_search = Button(window, text = 'Search', command = search_book)
btn_update = Button(window, text = 'Update Price', command = update_price)

lbl_name.grid(row = 0, column = 0)
name.grid(row = 0, column = 1)
lbl_category.grid(row = 1, column =0)
category.grid(row=1,column=1)
lbl_price.grid(row=2, column =0)
price.grid(row =2, column =1)
btn_add.grid(row =3, column=0)
btn_delete.grid(row=3, column=1)
btn_reset.grid(row=3, column=2)
btn_update.grid(row =3, column = 3)
search_name.grid(row=4, column =0)
btn_search.grid(row=4, column =1)

window.mainloop()
```
